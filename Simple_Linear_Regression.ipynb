```python
# Simple Linear Regression Example

# Step 1: Import libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Step 2: Sample data
# x = study hours, y = exam score
x = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)  # input feature
y = np.array([2, 4, 5, 4, 5])                 # target variable

# Step 3: Create model
model = LinearRegression()

# Step 4: Train (fit) the model
model.fit(x, y)

# Step 5: Predict for new values
x_new = np.array([[6]])
y_pred = model.predict(x_new)

print(f"Predicted score for 6 hours of study: {y_pred[0]:.2f}")

# Step 6: Visualization
plt.scatter(x, y, color='blue', label='Actual Data')
plt.plot(x, model.predict(x), color='red', label='Regression Line')
plt.xlabel("Study Hours")
plt.ylabel("Exam Score")
plt.title("Simple Linear Regression Example")
plt.legend()
plt.show()
```
