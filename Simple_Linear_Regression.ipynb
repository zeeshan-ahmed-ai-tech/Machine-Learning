```python
import pandas
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("/content/drive/MyDrive/Machine Learning/home_prices_pkr.csv")
print(df)
```

```python
plt.xlabel('Area_sqft')
plt.ylabel('Price_PKR')
plt.scatter(df.Area_sqft,df.Price_PKR,color='red',marker='+')
```

```python
new_df=df.drop('Price_PKR',axis='columns')
new_df
```

```python
model = linear_model.LinearRegression()
model.fit(df[['Area_sqft']], df.Price_PKR)
model.predict(pd.DataFrame({'Area_sqft': [2300]}))
```

```python
model.coef_
```

```python
model.intercept_
```

```python
df = pd.read_csv("/content/drive/MyDrive/Machine Learning/home_area_only.csv")
area_df = df[['Area_sqft']]
p = model.predict(area_df)
p
```

```python
area_df['Price_PKR']=p
area_df
area_df.to_csv('predection.csv')
```
